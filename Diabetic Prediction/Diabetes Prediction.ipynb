```python
#Data Visualization, Data Manipulation and Data Cleaning
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
#Machine learning Libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.isnull() #to check null values
```

```python
print(df.info())
```

```python
X= df.drop('Outcome', axis=1)#Split data into X features and target Y
Y=df['Outcome']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)#Training 80% data and testing 20% data
```

```python
model= DecisionTreeClassifier(max_depth=4, random_state=42)#Build and train the decision tree
model.fit(X_train, Y_train)
```

```python
Y_pred= model.predict(X_test)#Make prediction

print("Accuracy:", accuracy_score(Y_test, Y_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("Classificcation Report:\n", classification_report(Y_test, Y_pred))
```

```python
#Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X.columns, class_names=['No Diabetes', 'Diabetes'], filled=True)
plt.show()
```

```python
import pickle
#Save the Model
pickle.dump(model, open('diabetes_model.pkl', 'wb'))
print("Model saved as diabetes_model.pk1")
```
