```python
#Importing libraries for data visualization and manipulation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("student_habits_performance.csv")
```

```python
df.drop("student_id", axis=1, inplace=True)#Column not needed 
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder #Machine learning library
#Label Encode binary column
label_cols=['gender', 'part_time_job', 'extracurricular_participation']

le= LabelEncoder()
for col in label_cols:
    df[col]= le.fit_transform(df[col])

#Hot Encode Multi category
#df= -d.get_dummies(df, column=['diet_quality', 'exercise_frequency', 'parental_education_level', 'internet_quality'
#drop_first=True)
```

```python
df.columns
```

```python
X= df.drop('exam_score', axis=1)
Y= df['exam_score']
```

```python
X
```

```python
Y
```

```python
#Machine learning Tools
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model= LinearRegression()
model.fit(X_train, Y_train)
#Predict
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
#R2 score (how good the model is)
r2= r2_score(Y_test, Y_pred)
mse= mean_squared_error(Y_test, Y_pred)
rmse= np.sqrt(mse)

print(f"R2 Score: {r2:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sb.scatterplot(x=Y_test, y=Y_pred)
plt.xlabel("Actual Exam Score")
plt.ylabel("Pedicted Exam Score")
plt.title("Actual vs Predicted Scores")
plt.grid(True)
plt.show()
```
